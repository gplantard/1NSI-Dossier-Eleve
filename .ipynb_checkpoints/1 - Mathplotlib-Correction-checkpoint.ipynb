# Introduction à Matplotlib 

Nous allons voir les bases faire des représentation graphique de données. Une des bibliothèques les plus utilisée en Python pour cela est MathPlotLib

Voici un exemple de code pour afficher une ou plusieurs courbes.

#### Question : 
Modifiez ce code pour afficher les valeurs sur l'intervalle $[-3,3]$, en ajoutant la fonction $2x^2-5x-3$


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3, 3, 100) #crée une liste de 100 valeurs pour x allant de 0 à 2

fig,ax = plt.subplots()  # Crée une figure et un repère.
ax.plot(x, x, label='linéaire')  # Ajoute des données dans le repère (la fonction f(x)=x)
ax.plot(x, x**2, label='carré')  # Ajoute d'autre données dans le repère (la fonction g(x)=x²)
ax.plot(x, x**3, label='cube')  # la fonction h(x)=x^3
ax.plot(x,2*x**2-5*x-3, label="polynôme")
ax.set_xlabel('label pour x')  # Ajoute un label à l'axe des x.
ax.set_ylabel('label pour y')  # Ajoute un label à l'axe des y.
ax.set_title("graphique simple")  # Ajoute un titre au repère.
ax.legend()  # ajoute un légende (basée sur les "label" précédents).
plt.show()

print("type de x :",type(x))
print("type de x**2 :",type(x**2))
```

#### Question 
1. De quel type est x ?
2. De quel type est x**2 ?

Utilisez la fonction type(<objet>) pour vérifier votre réponse.

#### Réponse
1. C'est une liste / un tableau de valeurs
2. C'est égaleur une liste / un tableau de valeurs

*Commentaire* : On peut en déduire que pour afficher une courbe, MathplotLib prend en premier paramètre une liste de valeur pour x, puis en deuxième paramètre une liste de valeur de y correspondante.

# Affichage d'un nuage de point

La particularité de MathPlotLib est qu'il ne prend pas une liste de points avec pour chque point un x et un y ce qui paraîterai naturel.
Au lieu de ça il prend deux liste
* Une liste avec l'ensemble des x
* Une liste avec l'ensemble des y correspondant

Voici un exemple d'utilisation.

#### Question
Ajoutez dans le nuage les points suivants :
1. $(5;3)$
2. $(3;2,5)$
3. $(6,7;2,4)$
4. $(-3;\frac{5}{3})$


```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6,7,8,5,3  ,6.7,-3 ]
y = [4,1,3,6,1,3,5,2,3,2.5,2.4,5/3]

plt.scatter(x,y)

plt.title('Nuage de points avec Matplotlib')
plt.xlabel('x')
plt.ylabel('y')

#plt.savefig('ScatterPlot_01.png') #permet d'enregistrer la figure dans un fichier image
plt.show()
```

On peut également ajouter d'autres paramètres en plus des coordonnées comme la taille et la couleur.
#### Question
Modifier le code suivante pour changer la taille et la couleur du point de coordonnées (3;3)

```python
import matplotlib.pyplot as plt

x_list1 = [1,2,4] # on enlève le 3
y_list1 = [4,1,6] # on enlève le 3
size_list1 = [100,500,500] # on enlève le 200

plt.scatter(x_list1, y_list1, s=size_list1, c='coral')

x_list2 = [5,6,7,8,3] #on ajoute le 3
y_list2 = [1,3,5,2,3] #on ajoute le 3
size_list2 = [100,500,100,500, 200] #on ajoute le 200

plt.scatter(x_list2, y_list2, s=size_list2, c='lightblue')

plt.title('Nuage de points avec Matplotlib')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

#### Question
En vous aidant d'internet améliorez votre graphique avec les étapes suivantes:
* Changer la forme des points bleu
* Afficher l'axe des abscises et l'axe des ordonnées
* Afficher la grille (les axes verticaux et horizontaux)

```python
import matplotlib.pyplot as plt

x = [-3,1,2,3,4]
y = [-1,4,1,-3,6]
size = [100,500,100,500]

fig,ax = plt.subplots()  # Crée une figure et un repère.

ax.scatter(x, y, s=size, c='coral', marker="^") # Ajoute les points dans le repère

x = [5,6,7,8]
y = [1,3,5,2]
size = [100,500,100,500]

ax.scatter(x, y, s=size, c='lightblue', marker="*") # Ajoute les points dans le repère

#affichage de la grille
ax.grid(True,which="both", linestyle='--')

#récupération de valeur min et max en x et en y
(xmin, xmax) = ax.xaxis.get_view_interval()
(ymin, ymax) = ax.yaxis.get_view_interval()

# Affichage des axes des abscisses et ordonnées
ax.add_artist(plt.Line2D((0, 0), (ymin, ymax), linewidth = 1))
ax.add_artist(plt.Line2D((xmin, xmax), (0, 0), linewidth = 1))

ax.set_title('Nuage de points avec Matplotlib')
ax.set_xlabel('x')
ax.set_ylabel('y')

plt.show()
```
